## Find out null values in each column

```python
m=matches.isnull().sum()
df=pd.DataFrame(m)
x=df.index
y=df[0]

plt.figure(figsize=(10,10))
bars = plt.bar(x,y, width=.4)

# access the bar attributes to place the text in the appropriate location
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + 2.15, yval)
plt.xticks(rotation=90, fontsize=15)
plt.title('Null Values in each column', fontsize=20)
plt.yticks([])
plt.show()
```

## 1.Number of matches held season wise

```python

m=matches.groupby('season').id.count()
df=pd.DataFrame(m)
x=df.index
y=df['id']

plt.figure(figsize=(10,4))
bars = plt.bar(x,y, width=.5, color=('cyan'))

# access the bar attributes to place the text in the appropriate location
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + 0.001, yval, color='red')
plt.xticks(rotation=0, fontsize=15)
plt.title('Matches held in each season', fontsize=20)
plt.yticks([])
plt.show()
```

### More number of matches was held in Season 2013 : 76 Matches

## 2. Number of matches won by each team at each season

```python


m=matches.groupby(['season', 'winner'])['winner'].count()
#matches.groupby(['season'])['winner'].value_counts()
count=1
fig = plt.figure()
groups=m.groupby('season')
for year,group in groups:
    ax = fig.add_subplot(4,3,count)
    ax.set_title(year)
    ax=group[year].plot(kind="bar",figsize=(10,15),width=0.8, color=('green','red'))
    count=count+1
    plt.xlabel('')
    plt.yticks([])
    plt.ylabel('Matches Won', fontsize=10)
    
    totals = []
    for i in ax.patches:
        totals.append(i.get_height())
    total = sum(totals)
    for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
        ax.text(i.get_x()+0.2, i.get_height()-1.5,s= i.get_height(),color="white",fontweight='bold')
    
    
plt.tight_layout()
plt.show()
```

### Mumbai Indians and Chennai super kings has won maximum matches

## 3.Total matches played by each team

```python
totalmatches=matches.groupby('team1').id.count()+matches.groupby('team2').id.count()
#or matches['team1'].value_counts()+matches['team2'].value_counts()
df=pd.DataFrame(totalmatches)
x=df.index
y=df['id']
plt.figure(figsize=(14,6))
ax=sns.barplot(x=df.index,y=df['id'],palette="Set1")
plt.ylabel('Total number of matches', fontsize=15)
plt.xticks(rotation=90, fontsize=15)


totals = []
for i in ax.patches:
    totals.append(i.get_height())
total = sum(totals)
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+0.01, i.get_height()-0.001,s= i.get_height(),color="blue",fontweight='bold')
plt.show() 
```

## 4. How many matches won by each team ?

## 5.  Comparison between Number of matches won by each team and total matches played

```python


totalm=pd.DataFrame(totalmatches)
totalm
winningm=pd.DataFrame(winningmatches)
winningm
m=totalm['id'] - winningm['id']
df=pd.DataFrame(m)
x=df.index
y=df['id']
#print(x)
plt.figure(figsize=(10,4))
bars = plt.bar(x,y, width=.5, color=('pink'))

# access the bar attributes to place the text in the appropriate location
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + 0.001, yval, color='black')
plt.xticks(rotation=90, fontsize=15)
plt.title('Comparison between matches won and played by each team', fontsize=20)
plt.yticks([])
plt.show()
```

## 14.  Toss Decision across matches

```python

#matches.groupby('toss_decision')['id'].count()
toss=matches['toss_decision'].value_counts()
#m=pd.DataFrame(toss)
#print(m)
#plot = m.plot.pie(y='toss_decision', figsize=(5, 5))

toss=matches['toss_decision'].value_counts()
labels=np.array(toss.index)
sizes = toss.values
colors = ['gold', 'lightskyblue']

# Plot
plt.figure(figsize=(5,3))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)

plt.title('Toss Decision of all the matches')
plt.axis('equal')
plt.show()
```
